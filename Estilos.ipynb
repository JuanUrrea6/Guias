#Diferentes Estilos

```python
#Five-Thrity-Eight
t = linspace(0,100,50)
y1 = t**2
y2=2*t**2
coseno= 15000*abs(exp(-0.009*t)*np.cos(0.1*t))
```

```python
figure(figsize=(10,5))
with plt.style.context('fivethirtyeight'):
    plot(t,y1,r'v')
    plot(t,y2,r'bs')
    plot(t,coseno,r'-*',ms=12)
title('Estilo: Five-Thirty-Eight')
xlabel('t', color = 'Blue')
ylabel('Gr√°fica', position=(0,0.55), color = 'Red',style='italic')
```

```python
#Im√°genes aleatorias para Python
figure(figsize=(8,5))
eqs = []
eqs.append((r"$\nabla^2 V = -\rho/\epsilon_0$"))
eqs.append((r"$\nabla^2 V= 0$"))
eqs.append((r"$\nabla \cdot B = 0$"))
eqs.append((r"$-i\hbar \partial \psi / \partial t = \hbar^2 / 2m\nabla^2 \psi + V \psi$"))
eqs.append((r"$F_G = G\frac{m_1m_2}{r^2}$"))

axes([0.025, 0.025, 0.95, 0.95])

for i in range(15):
    index = np.random.randint(0, len(eqs))
    eq = eqs[index]
    size = np.random.uniform(12, 32)
    x,y = np.random.uniform(0, 1, 2)
    alpha = np.random.uniform(0.25, .75)
    text(x, y, eq, ha='center', va='center', color="#11557c", alpha=alpha,transform=gca().transAxes, 
    fontsize=size, clip_on=True)
xticks(())
yticks(())

show()
```
