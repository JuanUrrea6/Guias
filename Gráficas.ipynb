##Otras Formas de Mostrar Datos

```python
figure(figsize=(15,5))
subplot2grid((1,3), (0,0)) #Primer scatter plot
scatter(x,y, label = "Datos")
title("Simple")
legend(loc = 'upper left')
ylim(0, 120)
xlim(0,12)
xlabel("X")
ylabel("Y")

#Crear errores aleatorios pero pequeÃ±os
error = []
for i in range(20):
    actual = uniform(0,7)
    error.append(actual)

subplot2grid((1,3), (0,1))
colores = rand(10)
area = np.pi * (10*colores)**2 #Distintos tamaÃ±os
plot(x,y*1.4+10,'o--',label="Datos", markerfacecolor = "Red", markeredgecolor = "Blue", ms = 5, markeredgewidth = 1)
errorbar(x,y*1.4+10,yerr=error, ls="-o", lw = 2) #Barras de Error
title("EstÃ©tica Mejorada")
ylim(0,160)
xlim(0,12)
xlabel("X")
ylabel("Y")
legend(loc = "upper left")
grid(True)

xkcd()
subplot2grid((1,3), (0,2))
scatter(x,y,marker='*', s = area, c = "Green")
title("Experimentando")
ylim(0,120)
xlim(0,12)
xlabel("X")
ylabel("Y")
grid(True)
rcdefaults()
show()
```

```python
n = array([0,1,2,3,4,5]) #Arreglo de prueba nuevo
x2 = linspace(0,1,1000)
y2 = sin(4 * np.pi * x2) * exp(-5 * x2)
```

```python
figure(figsize=(15,13))
subplot2grid((2,3), (0,0))
bar(n, n**2, align="center", width=0.5, alpha=0.5)
xlim(0,6)
title("Diagrama de Barras")
xlabel("n")
ylabel("$n^2$")

subplot2grid((2,3), (0,1))
step(n, n**2, lw=2)
title("Diagrama escalonado")
xlabel("n")
ylabel("$n^2$")

subplot2grid((2,3), (0,2))
fill_between(n, n**2, n**3, color="Blue", alpha=0.5)
title("Ãrea entre Curvas")

subplot2grid((2,3), (1,0))
fill(x2,y2)

subplot2grid((2,3),(1,1))
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X + Y**2
V = 1 + X**2 - Y
speed = np.sqrt(U**2 + V**2)

streamplot(X, Y, U, V, color=U, linewidth=2, cmap='gist_heat')
colorbar()
title("Diagrama de PosiciÃ³n\n segÃºn velocidades")
show()
```
